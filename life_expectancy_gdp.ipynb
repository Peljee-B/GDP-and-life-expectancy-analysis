```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load the data
all_data = pd.read_csv('all_data.csv')
print(all_data.head())
```

```python
data_statistics = all_data.describe(include='all')
print(data_statistics)
```

```python
life_expectancy_mean = all_data.groupby('Country')['Life expectancy at birth (years)'].mean()
print(life_expectancy_mean)
```

```python
GDP_mean = all_data.groupby('Country')['GDP'].mean()
print(GDP_mean)
```

```python
# Line plot for Life Expectancy over the years
plt.figure(figsize=(12, 6))
sns.lineplot(data=all_data, x='Year', y='Life expectancy at birth (years)', hue='Country')
plt.title('Life Expectancy at Birth over years by Country')
plt.xlabel('Year')
plt.ylabel('Life Expectancy at Birth (years)')
plt.legend(title='Country')
plt.show()
```

```python
# Line plot for GDP over the years
plt.figure(figsize=(12, 6))
sns.lineplot(data=all_data, x='Year', y='GDP', hue='Country')
plt.title('GDP over years by Country')
plt.xlabel('Year')
plt.ylabel('GDP (USD)')
plt.legend(title='Country')
plt.show()
```

```python
# Scatter plot for GDP vs Life Expectancy
plt.figure(figsize=(12, 6))
sns.scatterplot(data=all_data, x='GDP', y='Life expectancy at birth (years)', hue='Country')
plt.title('GDP vs Life Expectancy at Birth')
plt.xlabel('GDP')
plt.ylabel('Life Expectancy at Birth (years)')
plt.legend(title='Country')
plt.show()
```

As GDP grows, life expectancy has been increasing to a certain level for all countries except Zimbabwe.
